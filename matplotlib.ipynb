## Bar Chart ##

```python
## Vertical Bar Chart

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King'Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232] # in $M

plt.bar(years, revenue, color="blue")
plt.title("Revenue in each Year")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")

# label the bars
for i in range(len(years)):
    plt.text(years[i], revenue[i]+10, str(revenue[i]), ha="center")

# increase size of graph
# y axis
plt.ylim(0, max(revenue)+ 100)
```

```python
## Bar Chart with multiple datasets
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King'Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M


non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label = "Oscar Movies") # Shifting this to LEFT
plt.bar(x+width/2, non_oscar_revenue, width, label = "NOn-Oscar Movies") # Shifting this to RIGHT

plt.title("Oscar VS Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue ($M)")

plt.legend()

plt.xticks(x, years)
```

## Scatter Plot

```python
# Scatter Plot

people = ["Person A", "Person B" , "Person C", "Person D", "Person E", "Person F","Person G","Person H","Person I","Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]


colors = ["green" if x < 135 else "red" for x in bp]


plt.scatter(age, bp,  s=bp, marker="o", alpha=1, cmap="viridis", c=bp)
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age VS BP")

plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy = (age[i], bp[i]), xytext=(age[i]+1, bp[i]+1))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)
```

```python
## Muliple Plots

cities = ["City A", "City B", "City C","City D", "City E"]

# Winter: Temperature Vs Humidity
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature VS Humidity
summer_temp = [35, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]


plt.scatter(winter_temp, winter_humidity, label= "Winter")
plt.scatter(summer_temp, summer_humidity, label= "Summer")

plt.title("Summer VS Winter Data")
plt.xlabel("Temperature")
plt.ylabel("Humidity")

plt.legend()
```

## Pie Charts


```python
# Pie Charts

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amount = [500, 150, 120, 100, 50]

explode = [0, 0, 0, 0.2, 0]

plt.style.use("default")
plt.pie(amount, labels=expenses, autopct="%1.1f%%",
        wedgeprops = {
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
            "fill" : True
        },
        explode = explode,
        startangle = 0,
        shadow = True
        )
plt.show()
```

## Histogram


```python
# Histogram

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

score = np.random.normal(70, 10, 100) # normal distribution 70-mean, 10-std dev

plt.hist(score, color="skyblue", edgecolor="black", bins=20)
plt.ylabel("Number of Student")
plt.xlabel("Scores of Students")
print(score)
```

```python
# custom data
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]

# custom bins
bins = [0, 30, 70, 90, 100]


plt.hist(
    marks,
    bins=bins,
    edgecolor="black",
    color="lightblue"
)
plt.axvline(
    x=33,
    label="Passing Marks (33)",
    color="red",
    linewidth="2",
    linestyle="--"
)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)
```

```python
# Multiple datasets

legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

# Transaction Amount Distribution - Legit VS Fraud
plt.hist(legit_transactions, color="green", bins=20, label="Legit Transaction", edgecolor="black", alpha= 0.5)
plt.hist(fraud_transactions, color="red", bins=20, label="Fraud Transaction", edgecolor="black", alpha= 0.5)

plt.title("Transaction Amount Disctribution - Legit VS Fraud")
plt.xlabel("Amount (in $M)")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
```
