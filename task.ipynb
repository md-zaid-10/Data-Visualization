# Weekly Temperature

Compare Temperatures of cities over a week using subplots and loops.

```python
import matplotlib.pyplot as plt


days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperature = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig, axes = plt.subplots(2, 2, figsize=(7,7))

# combinations: 0-0, 0-1, 1-0, 1-1

city_no = 0
for i in range(2): # represent rows
    for j in range(2): # represents columns
        axes[i][j].plot(days, temperature[city_no], marker='o')
        axes[i][j].set_title(cities[city_no])
        axes[i][j].grid(True)
        city_no = city_no + 1

fig.suptitle("Temperature in Cities iver the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")
fig.tight_layout()
```

```python
# another solution by flattening the index from 2-2 to 4

fig, axes = plt.subplots(2, 2, figsize=(7,7))
axes = axes.flatten()
for i, ax in enumerate(axes):
    ax.plot(days, temperature[i], marker="o")
    ax.set_title(cities[i])
    ax.grid(True)

fig.suptitle("Temperature in Cities iver the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")
fig.tight_layout()
```
